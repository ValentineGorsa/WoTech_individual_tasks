```python
# required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #visualisation
```

```python
pokemon_df = pd.read_csv('/content/sample_data/Pokemon.csv')
pokemon_df.head(n=10)
```

```python
pokemon_df.sample(5) #random 5 rows
```

```python
# How many rows and columns are in this data?

rows, columns = pokemon_df.shape
print(f'There are {rows} rows and {columns} columns.')
```

```python
# How many unique Pokemon names in this data?

unique_names = pokemon_df['Name'].nunique()
print(f'There are {unique_names} unique Pokemon names.')
print(unique_names)
```

```python
# How many pokemons have Attack greater than 100?

attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(f'There are {len(attack_greater_than_100)} pokemons with Attack greater than 100.')
```

```python
# How many pokemons Defense greater than 100?

defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(f'There are {len(defense_greater_than_100)} pokemons with Defense greater than 100.')
```

```python
# How many pokemons have both, attack and defense?

attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(f'There are {len(attack_defense_greater_than_100)} pokemons with both Attack and Defense greater than 100.')
```

```python
# VALUE COUNTS!!!
# Create a grouped DataFrame that counts the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df['Type 1'].value_counts().reset_index() # reset_index make indexex instead TYPE1 is index
print(type1_pokemons)
```

```python
# Bar chart

type1_pokemons.plot(x = 'Type 1', y = 'count', kind = 'bar', color='white', edgecolor='black')
plt.title('Number of Pokemons by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Number of Pokemons')
plt.show()
##
```

```python
# How to rotate x and y values?

plt.bar(type1_pokemons['Type 1'], type1_pokemons['count'], color=['thistle','skyblue'], edgecolor='maroon')
plt.xticks(rotation=90)
```

```python
# What is the maximum 'Speed' value? What is the minimum 'Speed' value?
# What is the difference between max and min 'Speed'?

max_speed = pokemon_df['Speed'].max()
print(f'The maximum Speed value is {max_speed}.')

min_speed = pokemon_df['Speed'].min()
print(f'The minimum Speed value is {min_speed}.')

difference_between_max_and_min_speed = max_speed - min_speed
print(f'The difference between max and min Speed is {difference_between_max_and_min_speed}.')
```

```python
# (DIFFICULT) Find Pokmon with the longest name (excluding spaces)?
# What is this pokemons name?

# Create a new column in the DataFrame with the length of the names excluding spaces
pokemon_df['NameLengthNoSpaces'] = pokemon_df['Name'].str.replace(" ", "").str.len()

# Find the index of the Pokmon with the longest name (excluding spaces)
longest_pokemon_name_idx = pokemon_df['NameLengthNoSpaces'].idxmax()

# Get the Pokmon's name using the index
longest_pokemon_name = pokemon_df.iloc[longest_pokemon_name_idx]["Name"]

print(f'The longest Pokmon name is {longest_pokemon_name}.')
```
