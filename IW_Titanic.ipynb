```python
## Exploratory Data Analysis (EDA)
import pandas as pd #analysis
import matplotlib.pyplot as plt #visualisation
titanic_df = pd.read_excel('/content/titanic3.xls')
```

```python
## Use Pandas methods to explore and understand the dataset:
# Display the first few rows to get an overview of the data structure.
titanic_df.head()
```

```python
# Check the dimensions of the dataset (number of rows and columns).
titanic_df.shape
```

```python
# Identify missing values and decide on strategies for handling them.
titanic_df.isnull().sum()
```

```python
# Compute basic statistics (e.g., mean, median, min, max) for numerical columns.
titanic_df.describe()
```

```python
max_age = titanic_df['age'].max()
min_age = titanic_df['age'].min()
mean_age = titanic_df['age'].mean()

print(f'Maximum age of Titanic passenger was {max_age}')
print(f'Minimum age of Titanic passenger was {min_age}')
print(f'Mean age of Titanic passenger was {mean_age}')
```

```python
# Analyze categorical variables (e.g., value counts)
titanic_df['sex'].value_counts()
```

```python
### Data Visualization
## Choose one column of the dataset that interests you
# Use Matplotlib pyplot to create a meaningful visualization:
# Select an appropriate plot type (e.g., bar chart, histogram, pie chart).

# Replace NaN values with a placeholder for visualization
titanic_df['age'] = titanic_df['age'].fillna(-1)

# Convert the age column back to numeric for sorting and plotting
titanic_df['age'] = pd.to_numeric(titanic_df['age'], errors='coerce')

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Titanic Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```
