```python
# Online shop prototype
# Define Classes
class Client:
    number_of_clients = 0

    def __init__(self, client_id, name):
        self.client_id = client_id
        self.name = name
        self.transactions = []
        Client.number_of_clients += 1

    def create_transaction(self, transaction_id, currency='EUR'):
        transaction = Transaction(transaction_id, self.client_id, currency)
        self.transactions.append(transaction)
        return transaction

class Item:
    def __init__(self, item_id, name, price, currency='EUR'):
        self.item_id = item_id
        self.name = name
        self.price = price
        self.currency = currency

class Transaction:
    def __init__(self, transaction_id, client_id, currency='EUR'):
        self.transaction_id = transaction_id
        self.client_id = client_id
        self.amount = 0
        self.currency = currency
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        self.amount += item.price

# Create instances of Clients
clients = [
    Client('1', 'Anna'),
    Client('2', 'Oskar'),
    Client('3', 'Jenifer'),
    Client('4', 'Rick')
]

# Create instances of Items
items = [
    Item('100', 'soap', 5),
    Item('101', 'shampoo', 20),
    Item('102', 'cream', 25),
    Item('103', 'balsam', 22),
    Item('104', 'tonic', 15)
]

# Create instances of Transactions and associate them with Clients
transactions = []

# Example: Create a transaction for Anna and add items to it
transaction1 = clients[0].create_transaction('T1', 'EUR')
transaction1.add_item(items[0])
transaction1.add_item(items[1])

# Example: Create a transaction for Oskar and add items to it
transaction2 = clients[1].create_transaction('T2', 'EUR')
transaction2.add_item(items[2])
transaction2.add_item(items[3])

# Example: Create a transaction for Jenifer and add items to it
transaction3 = clients[2].create_transaction('T3', 'EUR')
transaction3.add_item(items[4])

transaction4 = clients[3].create_transaction('T4', 'EUR')
transaction4.add_item(items[0])
transaction4.add_item(items[1])
transaction4.add_item(items[2])
transaction4.add_item(items[3])
transaction4.add_item(items[4])

# Add transactions to the list for easy access
transactions.extend([transaction1, transaction2, transaction3, transaction4])

# Print all clients, their transactions, and items in each transaction
print("All Clients:")
for client in clients:
    print(f'Number of clients: {Client.number_of_clients}.')
    print(f"Client ID: {client.client_id}, Name: {client.name}")
    print("Transactions:")
    for transaction in client.transactions:
        print(f"  Transaction ID: {transaction.transaction_id}, Amount: {transaction.amount} {transaction.currency}")
        print("  Items:")
        for item in transaction.items:
            print(f"    Item ID: {item.item_id}, Name: {item.name}, Price: {item.price} {item.currency}")
    print()
```
